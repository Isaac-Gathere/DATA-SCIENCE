```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("C:\\Users\\ADMIN\\Desktop\\Data-Analysis\\apartment_cost_list_20240317.csv")


```

```python
# Exploring
print("Dataset shape:", data.shape)
print("Columns:", data.columns)
print("\nData types:\n", data.dtypes)
print("\nFirst few rows:\n", data.head())

```

```python
data['Initial Cost'] = data['Initial Cost'].replace('[\$,]', '', regex=True).astype(float)

data = data.drop(columns=['Curb Cut', 'Horizontal Enlrgmt', 'Vertical Enlrgmt'])
data = data.dropna(subset=['Zoning Dist1', 'Fully Permitted'])

data.to_csv("cleaned_data.csv", index=False)

```

```python
print("\nMissing values:\n", data.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Borough', y='Initial Cost', data=data)
plt.title('Average Housing Costs Across Boroughs')
plt.xlabel('Borough')
plt.ylabel('Average Initial Cost')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(x='Job Type', data=data, order=data['Job Type'].value_counts().index)
plt.title('Distribution of Job Types')
plt.xlabel('Job Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
street_counts = data['Street Name'].value_counts()

# Get the top 10 streets with the most properties
top_streets = street_counts.head(10)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_streets, labels=top_streets.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Properties by Street')
plt.axis('equal')
plt.show()
```

```python
total_properties = top_streets.sum()

for street, count in top_streets.items():
    percentage = (count / total_properties) * 100
    print(f"{street}: {percentage:.1f}%")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Borough', y='Initial Cost', data=data)
plt.title('Distribution of Initial Costs Across Boroughs')
plt.xlabel('Borough')
plt.ylabel('Initial Cost')
plt.xticks(rotation=45)
plt.show()
```
